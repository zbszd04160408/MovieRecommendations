```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

import requests
from bs4 import BeautifulSoup
import timeit
import json
```

## Import main data

```python
main = pd.read_csv('Data/main_211110.csv', dtype=str)
main
```

```python
URL = "https://www.imdb.com/title/tt0118866/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

rels = soup.find_all("a", class_="ipc-poster-card__title ipc-poster-card__title--clamp-2 ipc-poster-card__title--clickable")

for rel in rels:
    link = rel.get("href")
    title = rel.find("span").text
    
    print(title, link)
```

```python
unique_id = main['id'].unique()
len(unique_id)
```

```python
start = timeit.default_timer()

tid_likes = {}
for i in range(len(unique_id)):
    tid = unique_id[i]
    URL = "https://www.imdb.com/title/" + tid + "/"
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")

    rels = soup.find_all("a", class_="ipc-poster-card__title ipc-poster-card__title--clamp-2 ipc-poster-card__title--clickable")
    
    links = []
    for rel in rels:
        href = rel.get("href")
        links.append(re.search(r'\/(.*?)\/', href[6:]).group(1))
    
    tid_likes[tid] = links

stop = timeit.default_timer()

print('Time: ', stop - start)
```

```python
tid_likes
```

```python
name = ["recommendation" + str(i) for i in range(1,13)]
df = pd.DataFrame.from_dict(tid_likes, orient='index', columns= name)
df.reset_index(level=0, inplace=True)
df.rename(columns={'index':'id'}, inplace=True)
df.to_csv('Data/imdb_recommendations_tid.csv', index=False)
df
```

```python
start = timeit.default_timer()

tid_likes = {}
for i in range(len(unique_id)):
    tid = unique_id[i]
    URL = "https://www.imdb.com/title/" + tid + "/"
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")

    rels = soup.find_all("a", class_="ipc-poster-card__title ipc-poster-card__title--clamp-2 ipc-poster-card__title--clickable")
    
    titles = []
    for rel in rels:
        titles.append(rel.find("span").text)
    
    tid_likes[tid] = titles

stop = timeit.default_timer()

print('Time: ', stop - start)
```
